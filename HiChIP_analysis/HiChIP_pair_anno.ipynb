```r
CTCF <- read.table('all_ctcf.bed',header=F)
colnames(CTCF) <- c("chr","start","end")
CTCF <- makeGRangesFromDataFrame(CTCF)
```

```r
C2_pair_anno$index1  %>% str_split('_') -> C2index1
```

```r
C2_pair_anno$index2  %>% str_split('_') -> C2index2
```

```r
get_dis_to_ctcf <- function(data){
    data$index1 %>% str_split('_') -> C2index1
    data$index2 %>% str_split('_') -> C2index2

    as.data.frame(do.call(rbind, C2index1), stringsAsFactors = FALSE) %>%  
        dplyr::rename("chr" = "V1", "start" = "V2", "end"='V3') %>% makeGRangesFromDataFrame() -> C2index1

    as.data.frame(do.call(rbind, C2index2), stringsAsFactors = FALSE) %>%  
       dplyr::rename( "chr" = "V1", "start" = "V2", "end"='V3') %>% makeGRangesFromDataFrame() -> C2index2

    dis1 <- distanceToNearest(C2index1,CTCF)
    dis2 <- distanceToNearest(C2index2,CTCF)

    dis1 %>% as.data.frame() %>% .[,'distance'] -> dis11
    dis2 %>% as.data.frame() %>% .[,'distance'] -> dis21
    data$index1_to_ctcf <- dis11 
    data$index2_to_ctcf <- dis21 
    return(data)
 }
```

```r
C2_pair_anno <-get_dis_to_ctcf(C2_pair_anno)
```

```r
C3_pair_anno <-get_dis_to_ctcf(C3_pair_anno)
```

```r
all_pair_anno <-get_dis_to_ctcf(all_pair_anno)
```

```r
join_and_rbind_peaks <- function(mGC1_tmp1, mGC1_tmp2) {
  # left and right ->统一成 chr/start/end
  tmp2_L <- mGC1_tmp2[, 1:3, drop = FALSE]
  colnames(tmp2_L) <- c("chr", "start", "end")

  tmp2_R <- mGC1_tmp2[, 4:6, drop = FALSE]
  colnames(tmp2_R) <- c("chr", "start", "end")

  mGC1_tmp2_joined <- rbind(tmp2_L, tmp2_R)

  # 保证 mGC1_tmp1 也是 chr/start/end（如果本来就是就无影响）
  tmp1 <- mGC1_tmp1[, c("chr","start","end"), drop = FALSE]

  # rbind 要求列完全一致
  if (!identical(colnames(tmp1), colnames(mGC1_tmp2_joined))) {
    stop("Column names do not match for rbind.")
  }

  res <- rbind(tmp1, mGC1_tmp2_joined)
  rownames(res) <- NULL
  unique(res)
}
```

```r
C2_tmpmerge <- join_and_rbind_peaks(C2_tmp1, C2_tmp2)
```

```r
C3_tmpmerge <- join_and_rbind_peaks(C3_tmp1, C3_tmp2)
```
