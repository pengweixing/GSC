```r
C2_pair_anno  = readRDS(file = "C2_pair_anno.Rdata")
C3_pair_anno = readRDS(file = "C3_pair_anno.Rdata")
all_pair_anno = readRDS(file = "All_pair_anno.Rdata")
```

```r
C3genes <- read.table('/disk1/pengweixing/Glio_cuttag/22.footprint/01.human/04.GBM_survival/C3.kegggenes.txt',header = F)
```

```r
C2genes <- read.table('/disk1/pengweixing/Glio_cuttag/22.footprint/01.human/04.GBM_survival/C2.kegggenes.txt',header = F)
```

```r
C2genes <- C2genes$V1
C3genes <- C3genes$V1
```

```r
colnames(C3_pair_anno)
```

```r
C3_pair_anno  %>% filter(GeneSym %in% C3genes)  %>% select(chr_L,start_L,end_L,chr_R,start_R,end_R,
                                                          Recurrence,Anno_L,Dis_L,Anno_R,Dis_R,GeneSym) -> C3_pair_anno2
```

```r
C2_pair_anno  %>% filter(GeneSym %in% C2genes)  %>% select(chr_L,start_L,end_L,chr_R,start_R,end_R,
                                                          Recurrence,Anno_L,Dis_L,Anno_R,Dis_R,GeneSym) -> C2_pair_anno2
```

```r
chr3			
```

```r
C2_enh = data.frame()
for(i in 1:nrow(C2_pair_anno2)){
    Dis_L = C2_pair_anno2[i,'Dis_L']
    Dis_R = C2_pair_anno2[i,'Dis_R']
    if(abs(Dis_L)>abs(Dis_R)){
        C2_pair_anno2[i,c('chr_L','start_L','end_L')] -> temp
        colnames(temp) <- c('chr','start','end')
        C2_enh = rbind(C2_enh,temp)
    }else{
        C2_pair_anno2[i,c('chr_R','start_R','end_R')] -> temp
        colnames(temp) <- c('chr','start','end')
        C2_enh = rbind(C2_enh,temp)
    }
    
}
```

```r
C3_enh = data.frame()
for(i in 1:nrow(C3_pair_anno2)){
    Dis_L = C3_pair_anno2[i,'Dis_L']
    Dis_R = C3_pair_anno2[i,'Dis_R']
    if(abs(Dis_L)>abs(Dis_R)){
        C3_pair_anno2[i,c('chr_L','start_L','end_L')] -> temp
        colnames(temp) <- c('chr','start','end')
        C3_enh = rbind(C3_enh,temp)
    }else{
        C3_pair_anno2[i,c('chr_R','start_R','end_R')] -> temp
        colnames(temp) <- c('chr','start','end')
        C3_enh = rbind(C3_enh,temp)
    }
    
}
```

```r
C2_enh <- unique(C2_enh)
```

```r
C3_enh <- unique(C3_enh)
```

```r
write.table(x = C2_enh,file = "C2_kegg_enhs.bed",quote = F,sep = "\t",row.names = F,col.names = F)
```

```r
write.table(x = C3_enh,file = "C3_kegg_enhs.bed",quote = F,sep = "\t",row.names = F,col.names = F)
```

```r
C3_enh  %>% reframe(chr=chr,start2=as.integer((start+end)/2),end = as.integer((start+end)/2)+1) -> C3_all_enh2
```

```r
C2_enh  %>% reframe(chr=chr,start2=as.integer((start+end)/2),end = as.integer((start+end)/2)+1) -> C2_all_enh2
```
